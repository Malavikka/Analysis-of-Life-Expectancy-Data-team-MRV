```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('Life_Expectancy_Data.csv')
```

```python
data=data.rename(index=str,columns={"Life_expectancy ":"Expectancy"})
```

```python
data.groupby('Status').Expectancy.plot(kind='kde')
plt.xlabel("Life_Expectancy")
plt.legend()
```
